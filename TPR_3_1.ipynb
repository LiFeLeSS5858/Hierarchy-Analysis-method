Относительные веса критериев R, M и A могут быть определены путем деления элементов каждого столбца на сумму элементов этого же столбца. Искомые относительные веса критериев вычисляются теперь в виде средних значений элементов соответствующих строк нормализованной матрицы A. Следовательно,

```python
import numpy as np 
from scipy. stats import gmean

A = np.array([[1,1,1/4],[1,1,1/5],[4,5,1]])
A1= A[0,0]+A[1,0]+A[2,0]
A2= A[0,1]+A[1,1]+A[2,1]
A3= A[0,2]+A[1,2]+A[2,2]
ASum= np.array([A1,A2,A3])
print("Матрица сравнения:")
print(A)
print("Суммы элементов столбцов:",ASum)
N = np.array([[1/A1,1/A2,(1/4)/A3],[1/A1,1/A2,(1/5)/A3],[4/A1,5/A2,1/A3]])
print("Матрица средних значений элементов:")
print(N)
Wc=(N[0,0]+N[0,1]+N[0,2])/3
Wo=(N[1,0]+N[1,1]+N[1,2])/3
Wp=(N[2,0]+N[2,1]+N[2,2])/3
W= np.array([Wc,Wo,Wp])
print("Среднее значение элементов строк:",W)


y1=gmean([1, 1, 1/4] )
y2=gmean([1,1,1/5] )
y3=gmean([4,5,1] )

Sumy = y1+y2+y3

yh1=y1/Sumy
yh2=y2/Sumy
yh3=y3/Sumy

Sumyh = yh1+yh2+yh3

a1 = yh1*A1
a2 = yh2*A2
a3 = yh3*A3


Suma= a1+a2+a3


IC = (Suma -3)/2
CC = 1.98*(3-2)/3
OC = IC/CC

print("Согласованность данных:",OC)
```

Относительные веса альтернативных решений, вычисляются в пределах каждого критерия H и R с использованием следующих двух матриц сравнения.

При делении элементов каждого столбца полученных матриц на сумму элементов этих же столбцов получаем следующие нормализованные матрицы.

```python
Ak= np.array([[1,2],[1/2,1]])
Ak1= Ak[0,0]+Ak[1,0]
Ak2= Ak[0,1]+Ak[1,1]
AkSum= np.array([Ak1,Ak2])
Nk = np.array([[(Ak[0,0]/AkSum[0]),(Ak[0,1]/AkSum[1])],[(Ak[1,0]/AkSum[0]),(Ak[1,1]/AkSum[1])]])
Wkh=(Nk[0,0]+Nk[0,1])/2
Wkp=(Nk[1,0]+Nk[1,1])/2
Wk= np.array([Wkh,Wkp])
print("Матрица сравнения Ak:")
print(Ak)
print("Суммы элементов столбцов:",AkSum)
print("Матрица средних значений элементов:")
print(Nk)
print("Среднее значение элементов строк:",Wk)
```

##Часть 2: Метод анализа иерархий

Метод анализа иерархий (МАИ) - математический инструмент системного подхода к сложным проблемам принятия решений. МАИ позволяет понятным и рациональным образом структурировать сложную проблему принятия решений в виде иерархии, сравнить и выполнить количественную оценку альтернативных вариантов решения.

image.png

```python
cr_amount = int(input("Введите кол-во критериев: "))
# просто итератор
i = 1
#создание матрицы 
matr = np.eye(cr_amount)
# цикл для заполнения матрицы
for q in range(i, cr_amount+1):
  for p in range(i+1, cr_amount+1):
    matr[q-1][p-1] = round(float(input("Введите сравнение критерия {0}-{1}: ".format(q, p))), 3)
    matr[p-1][q-1] = round((matr[q-1][p-1])**(-1), 2)
  i += 1
# создание списка весовых коэффициентов
sum_matr = [round(sum(j),2) for j in matr]
# вывод списка
for k in sum_matr:
  print(round(k/sum(sum_matr), 2), end=' ')
```
